```python
import requests##使用requests

url = requests.get("https://www.must.edu.tw/")##使用requests的方法抓出網址中的內容 存到url
print(url.text)##印出url的文字
```

```python
import matplotlib.pyplot as plt ##使用matplotlib.pyplot 簡化 plt

x = [1, 2, 3] ##x陣列 1 . 2 . 3
y = [4, 5, 6] ##y陣列 4 . 5 . 4

plt.bar(x,y)##使用plt的方法產生長條圖
plt.show ##把使用plt產生的長條圖印出
```

```python
plt.pie(x,labels=y)##使用plt的方法產生圓形圖
plt.show ##把使用plt產生的圓形圖印出
```
